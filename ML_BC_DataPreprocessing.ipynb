```python
import pandas as pd
```

```python
file_path = 'BankChurners.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
# Removing specified columns
data = data.drop(columns=[
    "CLIENTNUM", 
    "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1", 
    "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"
])

# Display the updated dataframe structure
data.info()
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and temporary sets (60%, 40%)
train_data, temp_data = train_test_split(data, test_size=0.4, stratify=data['Attrition_Flag'], random_state=42)

# Splitting the temporary set into validation and test sets equally (50% each of 40% => 20% each of total)
validation_data, test_data = train_test_split(temp_data, test_size=0.5, stratify=temp_data['Attrition_Flag'], random_state=42)

# Checking the size of each dataset
(train_data.shape, validation_data.shape, test_data.shape)
```

```python
# Checking for missing values in each dataset
missing_values = {
    "Training Set": train_data.isnull().sum(),
    "Validation Set": validation_data.isnull().sum(),
    "Test Set": test_data.isnull().sum()
}

missing_values
```

```python
# Define placeholder values that might indicate missing data
placeholders = ["Unknown", "NA", "N/A", "?", "-"]

# Count placeholder values in each dataset
placeholder_counts = {
    "Training Set": train_data.isin(placeholders).sum(),
    "Validation Set": validation_data.isin(placeholders).sum(),
    "Test Set": test_data.isin(placeholders).sum()
}

placeholder_counts
```

```python
# Define columns to impute with the mode (most frequent category)
columns_to_impute = ['Education_Level', 'Marital_Status', 'Income_Category']

# Calculate the mode for each column from the training set
modes = train_data[columns_to_impute].mode().iloc[0]

# Impute 'Unknown' placeholders in training, validation, and test sets
for dataset in [train_data, validation_data, test_data]:
    for column in columns_to_impute:
        dataset[column] = dataset[column].replace('Unknown', modes[column])

# Check if 'Unknown' values are replaced by confirming non-existence in the datasets
replaced_check = {
    "Training Set": train_data[columns_to_impute].isin(['Unknown']).sum(),
    "Validation Set": validation_data[columns_to_impute].isin(['Unknown']).sum(),
    "Test Set": test_data[columns_to_impute].isin(['Unknown']).sum()
}

replaced_check
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Define columns for one-hot encoding and label encoding
nominal_columns = ['Gender', 'Marital_Status', 'Card_Category']
ordinal_columns = ['Education_Level', 'Income_Category']
target_column = 'Attrition_Flag'

# Set up the one-hot encoder and label encoders
onehot_encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' to avoid multicollinearity
label_encoder = LabelEncoder()

# Fit and transform the nominal columns with one-hot encoding
onehot_encoded_train = onehot_encoder.fit_transform(train_data[nominal_columns])
onehot_encoded_validation = onehot_encoder.transform(validation_data[nominal_columns])
onehot_encoded_test = onehot_encoder.transform(test_data[nominal_columns])

# Fit and transform the ordinal columns with label encoding
for column in ordinal_columns:
    label_encoder.fit(train_data[column])
    train_data[column] = label_encoder.transform(train_data[column])
    validation_data[column] = label_encoder.transform(validation_data[column])
    test_data[column] = label_encoder.transform(test_data[column])

# Encode the target variable
target_encoder = LabelEncoder()
train_data[target_column] = target_encoder.fit_transform(train_data[target_column])
validation_data[target_column] = target_encoder.transform(validation_data[target_column])
test_data[target_column] = target_encoder.transform(test_data[target_column])

# Convert one-hot encoded arrays back to DataFrame to concatenate with original data
onehot_columns = onehot_encoder.get_feature_names_out(nominal_columns)
train_onehot_df = pd.DataFrame(onehot_encoded_train, columns=onehot_columns, index=train_data.index)
validation_onehot_df = pd.DataFrame(onehot_encoded_validation, columns=onehot_columns, index=validation_data.index)
test_onehot_df = pd.DataFrame(onehot_encoded_test, columns=onehot_columns, index=test_data.index)

# Drop the original nominal columns and concatenate the one-hot encoded columns
train_data = pd.concat([train_data.drop(columns=nominal_columns), train_onehot_df], axis=1)
validation_data = pd.concat([validation_data.drop(columns=nominal_columns), validation_onehot_df], axis=1)
test_data = pd.concat([test_data.drop(columns=nominal_columns), test_onehot_df], axis=1)

# Verify the transformation by showing the new structure of the datasets
train_data.head()
```

```python
from imblearn.over_sampling import SMOTE

# Initialize the SMOTE sampler
smote = SMOTE(random_state=42)

# Separate features and target variable in training dataset
X_train = train_data.drop('Attrition_Flag', axis=1) 
y_train = train_data['Attrition_Flag']

# Apply SMOTE to the training data
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Convert the resampled features and target back into a DataFrame
X_train_smote_df = pd.DataFrame(X_train_smote, columns=X_train.columns)  # Ensuring column names are the same
y_train_smote_df = pd.DataFrame(y_train_smote, columns=['Attrition_Flag'])  # Creating a DataFrame for the target variable

# Concatenate the features and target variable into one DataFrame
balanced_train_data = pd.concat([X_train_smote_df, y_train_smote_df], axis=1)

balanced_train_data.head()
```

```python
# Calculate the sizes of the original and SMOTE-applied datasets
dataset_sizes = {
    "Training Set (Original)": train_data.shape,
    "Training Set (After SMOTE)": balanced_train_data.shape,  # adding target variable column count
    "Validation Set": validation_data.shape,
    "Test Set": test_data.shape
}

dataset_sizes
```

```python
# Print the class distribution in the balanced training set
print(y_train_smote.value_counts())
```

```python
from sklearn.preprocessing import RobustScaler

# Initialize the scaler
scaler = RobustScaler()

# List of numerical columns to scale
numerical_columns = [
    'Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count', 
    'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
    'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct', 
    'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio'
]

# Fit the scaler on the numerical columns of the training data (using SMOTE-applied data)
scaler.fit(balanced_train_data[numerical_columns])

# Transform the numerical columns in the training, validation, and test sets
balanced_train_data[numerical_columns] = scaler.transform(balanced_train_data[numerical_columns])
validation_data[numerical_columns] = scaler.transform(validation_data[numerical_columns])
test_data[numerical_columns] = scaler.transform(test_data[numerical_columns])
```

```python
# Ensure all dataframes have the same columns and order
columns = balanced_train_data.columns 
validation_data = validation_data.reindex(columns=columns)
test_data = test_data.reindex(columns=columns)
```

```python
# Check if all datasets have the same columns and column order
columns_match_train_validation = (balanced_train_data.columns == validation_data.columns).all()
columns_match_train_test = (balanced_train_data.columns == test_data.columns).all()

# Print the results
print("Do the training and validation datasets have the same columns and order?", columns_match_train_validation)
print("Do the training and test datasets have the same columns and order?", columns_match_train_test)
```

```python
balanced_train_data.to_csv('balanced_train_data.csv', index=False)
validation_data.to_csv('validation_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
```
